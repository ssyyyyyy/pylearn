## 温度图像

```python
import random
import matplotlib.pyplot as plt

x = range(60)
y = [random.uniform(15, 18) for i in x]
y1 = [random.uniform(20, 25) for i in x]
plt.figure(figsize=(20, 8), dpi=80)

plt.plot(x, y, label='Shanghai', color='r', linestyle='-.', linewidth=2)
plt.plot(x, y1, label='Beijing', color='b', linestyle='-', linewidth=2)
plt.yticks(range(0, 40, 5))
plt.xticks(x[::5])
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=10000)  # 修改这里
plt.xlabel('Time')
plt.grid(alpha=0.4,linestyle='--',color='r',which='major',linewidth=1.5)
plt.ylabel('Temperature')
plt.title('Temperature Variation')
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np

x = range(60)
y = [random.uniform(15, 18) for i in x]
y1 = [random.uniform(20, 25) for i in x]
y2 = [random.uniform(25, 30) for i in x]
y3 = [random.uniform(30, 35) for i in x]
figure, ax = plt.subplots(nrows=2, ncols=2, figsize=(20, 8), dpi=80)
ax[0, 0].plot(x, y, label='Shanghai', color='r', linestyle='-.', linewidth=2)
ax[0, 1].plot(x, y1, label='Beijing', color='b', linestyle='-', linewidth=2)
ax[1, 0].plot(x, y2, label='Guangzhou', color='g', linestyle='--', linewidth=2)
ax[1, 1].plot(x, y3, label='Shenzhen', color='y', linestyle=':', linewidth=2)
ax[0, 0].set_yticks(range(0, 40, 5))
ax[0, 1].set_xticks(x[::5])
ax[0, 0].set_xticks(x[::5])
ax[0, 1].set_yticks(range(0, 40, 5))
ax[0, 0].set_title('Temperature Variation')
ax[0, 1].set_xlabel('Time')
ax[1, 0].legend(loc='lower center', bbox_to_anchor=(0.5, -0.3),ncol=10000)
ax[1, 0].set_xlabel('Time')
ax[1, 0].set_ylabel('Temperature')
ax[0, 0].grid(alpha=0.4, linestyle='--', color='#aaa111', which='major', linewidth=1.5)
ax[0, 1].set_ylabel('Temperature')
ax[1, 0].set_title('Temperature Variation')

plt.show()
```
